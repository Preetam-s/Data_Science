```python
def outlier_data(col):
    sorted(col)
    Q1,Q3 = col.quantile([0.25,0.75])
    IQR = Q3-Q1
    LL = Q1 - (1.5*IQR)
    UL = Q3 + (1.5*IQR)
    return LL,UL
```

```python
outlier_data(df['Total_salary'])
```
